```python
error_count = 0
artwork_count = 0
with open('artwork_names1.json', 'a') as file:
    for i in range(df_original.shape[0]):
        #testing to see if /artists/{id} will generate more artist ids
        count_url = df_original['artwork_link'][i]
        artist = requests.get(count_url, headers = headers)
        try:
        #format json and load text
            j3 = json.loads(artist.text)
            if '_embedded' in j3.keys():
                if len(j3['_embedded']['artworks']) != 0:
                    try:
                        for work in j3['_embedded']['artworks']:
                            file.write(json.dumps(work) + '\n')
                            artwork_count += 1
                    except TypeError:
                        error_count += 1      
        except json.JSONDecodeError:
            error_count += 1
```
