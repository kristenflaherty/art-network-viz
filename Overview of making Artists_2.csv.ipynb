# Creating one CSV to store all data retrieved from Artsy API
## row = artist
## columns = artist feature

```python
import pandas as pd
import json
import numpy as np
import statistics as stats
```

```python
#original downloaded data
df = pd.read_json('ArtistSALL.json', lines = True)
```

```python
#before dups removed
print(df.shape[0], 'artists')
```

```python
#eliminate duplicates
df.drop_duplicates(subset=['id'], keep='last', inplace=True )
```

```python
#after dups removed
print(df.shape[0], 'artists')
```

```python
df.fillna(0, inplace= True)
df.replace({'':0}, inplace=True)
```

```python
df['nationality'] = df['nationality'].astype('str')
df['location'] = df['location'].astype('str')
df['hometown'] = df['hometown'].astype('str')
```

```python
#cleaning parts of nationality data
df['nationality'].replace({'America':'American'}, inplace=True)
df['nationality'].replace({'Africanamerican':'African american'}, inplace=True)
df['nationality'].replace({'American both':'American'}, inplace=True)
df['nationality'].replace({'United states':'American'}, inplace=True)
df['nationality'].replace({'Usa':'American'}, inplace=True)
df['nationality'].replace({'Us':'American'}, inplace=True)
df['nationality'].replace({'United kingdom':'British'}, inplace=True)
df['nationality'].replace({'Uk':'British'}, inplace=True)
df['nationality'].replace({'Venezuela':'Venezuelan'}, inplace=True)
df['nationality'].replace({'Ukraine':'Ukrainian'}, inplace=True)
df['nationality'].replace({'Wales':'Welsh'}, inplace=True)
df['nationality'].replace({'Uganda':'Ugandan'}, inplace=True)
df['nationality'].replace({'Uruguay':'Uruguayan'}, inplace=True)
df['nationality'].replace({'Ukranian':'Ukrainian'}, inplace=True)
```

```python
#reformatting links in url pages
df['artwork_link'] = [df.iloc[i]['_links'].get('artworks', {}).get('href', '') for i in range(df.shape[0])]
df['similar_artists'] = [df.iloc[i]['_links'].get('similar_artists', {}).get('href', '') for i in range(df.shape[0])]
df['genes'] = [df.iloc[i]['_links'].get('genes', {}).get('href', '') for i in range(df.shape[0])]
df['published_artworks'] = [df.iloc[i]['_links'].get('published_artworks', {}).get('href', '') for i in range(df.shape[0])]
df['similar_contemporary_artists'] = [df.iloc[i]['_links'].get('similar_contemporary_artists', {}).get('href', '') for i in range(df.shape[0])]
df.iloc[0]['_links']
```

```python
#create copy of semi-clean csv into new csv, with birthyear and deathyear as new columns
df = df[['id', 'sortable_name', 'birthday', 'deathday', 'gender', 'nationality', 'hometown', 'location', 'similar_artists', 'genes', 'artwork_link', 'published_artworks']].copy()
```

```python
#rename birthday column
df.rename(columns={'birthday':'birthyear'}, inplace=True)
```

```python
#rename deathday column
df.rename(columns={'deathday':'deathyear'}, inplace=True)
```

```python
#top safe dataframe to csv or hdf
df.to_csv('artists_full.csv', sep=',', header=True, index=False)
```

```python
df['nationality'] = df['nationality'].str.capitalize()
df['gender'] = df['gender'].str.capitalize()
```
